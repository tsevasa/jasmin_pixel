```python
import csv

csv_file = "_A_to_D_Intensity_Results.csv"
prefixes = ["A", "B", "C", "D"]
num_cells_cutoff = 50

counts = {
    p: {"active": 0, "inactive": 0}
    for p in prefixes
}

with open(csv_file, newline="") as f:
    reader = csv.DictReader(f)
    
    for row in reader:
        name = row["Name"]
        prefix = name[0]
        
        if prefix not in counts:
            continue
        
        num_active = int(row["NumActive"])
        num_inactive = int(row["NumInactive"])
        total_cells = num_active + num_inactive
        
        if total_cells >= num_cells_cutoff:
            counts[prefix]["active"] += num_active
            counts[prefix]["inactive"] += num_inactive

# Print results
for p in prefixes:
    a = counts[p]["active"]
    i = counts[p]["inactive"]
    print(f"{p}: Active percentage = {100*a/(a+i) :.1f}%, Active = {a}, Inactive = {i}, Total = {a + i}")
```

```python
import csv

csv_file = "_03_to_06_Intensity_Results.csv"
prefixes = ["A - 03", "A - 04", "A - 05", "A - 06",
            "B - 03", "B - 04", "B - 05", "B - 06"]
num_cells_cutoff = 5

counts = {
    p: {"active": 0, "inactive": 0}
    for p in prefixes
}

with open(csv_file, newline="") as f:
    reader = csv.DictReader(f)
    
    for row in reader:
        name = row["Name"]
        
        # Find matching prefix
        matched_prefix = None
        for p in prefixes:
            if name.startswith(p):
                matched_prefix = p
                break
        
        if matched_prefix is None:
            continue
        
        num_active = int(row["NumActive"])
        num_inactive = int(row["NumInactive"])
        total_cells = num_active + num_inactive
        
        if total_cells >= num_cells_cutoff:
            counts[matched_prefix]["active"] += num_active
            counts[matched_prefix]["inactive"] += num_inactive

# Print results
for p in prefixes:
    a = counts[p]["active"]
    i = counts[p]["inactive"]
    if a + i != 0:
        print(
            f"{p}: Active percentage = {100*a/(a+i):.1f}%, "
            f"Active = {a}, Inactive = {i}, Total = {a + i}"
        )
        
```
